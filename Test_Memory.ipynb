```python
opened_file = open('w1/g.csv')
from csv import reader
import numpy as np

read_file = reader(opened_file)
apps_data = list(read_file)
myarray = np.asarray(apps_data)
A1 = []
A2 = []
A3 = []
A4 = []
for i in range(0,50):
    A1.append(myarray[i][0])
    A2.append(myarray[i][1])
    A3.append(myarray[i][2])
    A4.append(myarray[i][3])

B1 = []
B2 = []
B3 = []
B4 = []
for j in range(0,2):
    for i in range(0,25):
        B1.append(myarray[i][j])
for j in range(2,4):
    for i in range(0,25):
        B2.append(myarray[i][j])
for j in range(0,2):
    for i in range(25,50):
        B3.append(myarray[i][j])
for j in range(2,4):
    for i in range(25,50):
        B4.append(myarray[i][j])
```

```python
import random
chu = input('Chon bai tap: ')
if chu == 'A1':
    a = A1
if chu == 'A2':
    a = A2
if chu == 'A3':
    a = A3
if chu == 'A4':
    a = A4
if chu == 'B1':
    a = B1
if chu == 'B2':
    a = B2
if chu == 'B3':
    a = B3
if chu == 'B4':
    a = B4
diem = 0
for k in range(10):
    b1 = random.randint(0, 25)
    if b1 == 1:
        b0 = a[0]
    else:
        b0 = a[b1 - 1]
    if b1 == 25:
        b2 = a[24]
    else:
        b2 = a[b1 + 1]
    print('tu: ', a[b1])
    s0 = input(('Dien tu truoc tu ', a[b1] ))
    s2 = input(('Dien tu sau tu ', a[b1] ))
    if s0 == b0:
        diem += 1
        print('dung truoc')
    else:
        print('sai truoc')
    if s2 == b2:
        diem += 1
        print('dung sau')
    else:
        print('sai sau')
    print('\n')
print('Tong ket ', diem, '/20')
```
